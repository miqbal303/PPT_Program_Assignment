Given an array, for each element find the value of the nearest element to the right which is having a frequency greater than that of the current element. If there does not exist an answer for a position, then make the value â€˜-1â€™.

**Examples:**
Input: a[] = [1, 1, 2, 3, 4, 2, 1]
Output : [-1, -1, 1, 2, 2, 1, -1]

Explanation:
Given array a[] = [1, 1, 2, 3, 4, 2, 1]
Frequency of each element is: 3, 3, 2, 1, 1, 2, 3

Lets calls Next Greater Frequency element as NGF
1. For element a[0] = 1 which has a frequency = 3,
   As it has frequency of 3 and no other next element
   has frequency more than 3 so  '-1'
2. For element a[1] = 1 it will be -1 same logic
   like a[0]
3. For element a[2] = 2 which has frequency = 2,
   NGF element is 1 at position = 6  with frequency
   of 3 > 2
4. For element a[3] = 3 which has frequency = 1,
   NGF element is 2 at position = 5 with frequency
   of 2 > 1
5. For element a[4] = 4 which has frequency = 1,
   NGF element is 2 at position = 5 with frequency
   of 2 > 1
6. For element a[5] = 2 which has frequency = 2,
   NGF element is 1 at position = 6 with frequency
   of 3 > 2
7. For element a[6] = 1 there is no element to its
   right, hence -1
    
    
Input : a[] = [1, 1, 1, 2, 2, 2, 2, 11, 3, 3]
Output : [2, 2, 2, -1, -1, -1, -1, 3, -1, -1]    
</aside>

```python
def find_nearest_higher_freq(arr):
    """
    Finds the nearest higher frequency element for each element in the array.
  
    Args:
      arr (list): The input array.
  
    Returns:
      list: The output array with the nearest higher frequency element for each element.
    """
  
    # Initialize an empty dictionary to store frequencies.
    freq = {}
    # Loop through the array and update frequencies.
    for x in arr:
      freq[x] = freq.get(x, 0) + 1
  
    # Initialize an empty stack to store indices.
    stack = []
    # Initialize an output array with -1 values.
    output = [-1] * len(arr)
  
    # Loop through the array from right to left.
    for i in range(len(arr) - 1, -1, -1):
      # Pop the stack until it is empty or the top element has a higher frequency than the current element.
      while stack and freq[arr[stack[-1]]] <= freq[arr[i]]:
        stack.pop()
  
      # If the stack is not empty, update the output array with the value of the top element of the stack.
      if stack:
        output[i] = arr[stack[-1]]
  
      # Push the current index to the stack.
      stack.append(i)
  
    # Return the output array.
    return output

# test the code with an example input
arr = [1, 1, 1, 2, 2, 2, 2, 11, 3, 3]
print(find_nearest_higher_freq(arr))
```

ðŸ’¡ **Question 2**

Given a stack of integers, sort it in ascending order using another temporary stack.

**Examples:**
Input : [34, 3, 31, 98, 92, 23]
Output : [3, 23, 31, 34, 92, 98]

Input : [3, 5, 1, 4, 2, 8]
Output : [1, 2, 3, 4, 5, 8]
</aside>

```python
def sort_stack(input_stack):
    """
    Sorts a stack using the following algorithm:
  
    1. Initialize an empty output stack.
    2. Loop until the input stack is empty.
    3. Pop the top element from the input stack and store it in a temporary variable.
    4. While the output stack is not empty and the top element of the output stack is greater than the temporary variable:
      * Pop the top element from the output stack and push it to the input stack.
    5. Push the temporary variable to the output stack.
    6. Return the output stack as the final result.
  
    Args:
      input_stack (list): The input stack.
  
    Returns:
      list: The sorted output stack.
    """
  
    # Initialize an empty output stack.
    output_stack = []
  
    # Loop until the input stack is empty.
    while input_stack:
  
      # Pop the top element from the input stack and store it in a temporary variable.
      temp = input_stack.pop()
  
      # While the output stack is not empty and the top element of the output stack is greater than the temporary variable:
      while output_stack and output_stack[-1] > temp:
  
        # Pop the top element from the output stack and push it to the input stack.
        input_stack.append(output_stack.pop())
  
      # Push the temporary variable to the output stack.
      output_stack.append(temp)
  
    # Return the output stack as the final result.
    return output_stack


# test the code with an example input
input_stack =  [34, 3, 31, 98, 92, 23]
print(sort_stack(input_stack))
```

Given a stack withÂ **push()**,Â **pop()**, andÂ **empty()**Â operations, The task is to delete theÂ **middle**Â elementÂ ****of it without using any additional data structure.

Input Â :Â Stack[] = [1, 2, 3, 4, 5]

Output :Â Stack[] = [1, 2, 4, 5]

Input Â :Â Stack[] = [1, 2, 3, 4, 5, 6]

Output :Â Stack[] = [1, 2, 4, 5, 6]

</aside>

```python
def delete_middle(stack, count):
    """
    Deletes the middle element from a stack using the following algorithm:
  
    1. If the stack is empty, return.
    2. Pop the top element from the stack and store it in a temporary variable.
    3. Call the helper function recursively with the stack and the count decremented by one.
    4. If the count is not zero, push the temporary variable back to the stack.
  
    Args:
      stack (list): The input stack.
      count (int): The number of elements to delete from the stack.
  
    Returns:
      list: The modified stack as the final result.
    """
  
    # Base case: if the stack is empty, return
    if not stack:
      return
  
    # Pop the top element from the stack and store it in a temporary variable.
    temp = stack.pop()
  
    # Call the helper function recursively with the stack and the count decremented by one.
    delete_middle(stack, count - 1)
  
    # If the count is not zero, push the temporary variable back to the stack.
    if count != 0:
      stack.append(temp)


def delete_middle_element(stack):
    """
    Deletes the middle element from a stack.
  
    Args:
      stack (list): The input stack.
  
    Returns:
      list: The modified stack as the final result.
    """
  
    # Call the helper function with the stack and its size divided by two as arguments.
    delete_middle(stack, len(stack) // 2)
  
    # Return the modified stack as the final result.
    return stack


# test the code with an example input
stack = [1, 2, 3, 4, 5]
print(delete_middle_element(stack))
```

Given a Queue consisting of firstÂ **n**Â natural numbers (in random order). The task is to check whether the given Queue elements can be arranged in increasing order in another Queue using a stack. The operation allowed are:

1. Push and pop elements from the stack
2. Pop (Or Dequeue) from the given Queue.
3. Push (Or Enqueue) in the another Queue.

**Examples :**

Input :Â Queue[] = { 5, 1, 2, 3, 4 }Â 

Output :Â YesÂ 

Pop the first element of the given Queue 

i.e 5.Â Push 5 into the stack.Â 

Now, pop all the elements of the given Queue and push them toÂ second Queue.Â 

Now, pop element 5 in the stack and push it to the second Queue.Â Â Â 

Input :Â Queue[] = { 5, 1, 2, 6, 3, 4 }Â 

Output :Â NoÂ 

Push 5 to stack.Â 

Pop 1, 2 from given Queue and push it to another Queue.Â 

Pop 6 from given Queue and push to stack.Â 

Pop 3, 4 from given Queue and push to second Queue.Â 

Now, from using any of above operation, we cannot push 5Â into the second Queue because it is below the 6 in the stack.

</aside>

```python
def can_arrange_increasing(queue):
    """
    Checks if the elements in the input queue can be arranged in increasing order using a stack.
  
    Args:
      queue (list): The input queue.
  
    Returns:
      str: "Yes" if the elements can be arranged in increasing order, "No" otherwise.
    """
  
    # Initialize an empty stack and an empty second queue.
    stack = []
    second_queue = []
  
    # Loop through the input queue.
    for num in queue:
  
      # While the stack is not empty and the top element of the stack is less than the current number, pop the top element from the stack and push it to the second queue.
      while stack and stack[-1] < num:
        second_queue.append(stack.pop())
  
      # Push the current number to the stack.
      stack.append(num)
  
    # While the stack is not empty, pop the top element from the stack and push it to the second queue.
    while stack:
      second_queue.append(stack.pop())
  
    # Check if the elements in the second queue are in increasing order.
    for i in range(len(second_queue) - 1):
      if second_queue[i] > second_queue[i + 1]:
        return "No"
  
    # Return "Yes" if the elements in the second queue are in increasing order.
    return "Yes"




# Test example 1
queue1 = [5, 1, 2, 3, 4]
print(can_arrange_increasing(queue1)) 

# Test example 2
queue2 = [5, 1, 2, 6, 3, 4]
print(can_arrange_increasing(queue2)) 
```

ðŸ’¡ **Question 5**

Given a number , write a program to reverse this number using stack.

**Examples:**
Input : 365
Output : 563

Input : 6899
Output : 9986
</aside>

```python
def reverse_number(number):
    """
    Reverses a number using a stack.
  
    Args:
      number (int): The number to be reversed.
  
    Returns:
      int: The reversed number.
    """
  
    # Initialize an empty stack and an empty string.
    stack = []
    string = ""
  
    # Convert the number into a string.
    number = str(number)
  
    # Loop through each character of the string and push it into the stack.
    for char in number:
      stack.append(char)
  
    # Loop until the stack is empty and pop each character from the stack.
    while stack:
      # Append it to the string.
      string += stack.pop()
  
    # Convert the string back to a number and return it.
    return int(string)


# test the code with an example input
number = 365
print(reverse_number(number))
```

Given an integer k and aÂ **[queue](https://www.geeksforgeeks.org/queue-data-structure/)**Â of integers, The task is to reverse the order of the firstÂ **k**Â elements of the queue, leaving the other elements in the same relative order.

Only following standard operations are allowed on queue.

- **enqueue(x) :**Â Add an item x to rear of queue
- **dequeue() :**Â Remove an item from front of queue
- **size() :**Â Returns number of elements in queue.
- **front() :**Â Finds front item.
</aside>

```python
from queue import Queue

def reverse_first_k_elements(queue, k):
    """
    Reverses the order of the first k elements of the input queue using a stack.

    Args:
      queue (Queue): The input queue.
      k (int): The number of elements to be reversed.

    Returns:
      Queue: The reversed queue.
    """

    # Initialize an empty stack.
    stack = []

    # Loop through the first k elements of the queue and push them onto the stack.
    for i in range(k):
        stack.append(queue.get())

    # Enqueue the elements from the stack into the queue, effectively reversing their order.
    while stack:
        queue.put(stack.pop())

    # Return the reversed queue.
    return queue

# Test the code with an example input
queue = Queue()
queue.put(10)
queue.put(20)
queue.put(30)
queue.put(40)
queue.put(50)
queue.put(60)
queue.put(70)
queue.put(80)
queue.put(90)
queue.put(100)

k = 5

other_queue = reverse_first_k_elements(queue, k)

# Print the elements of the other queue
while not other_queue.empty():
    print(other_queue.get(), end=" ")
```

Given a sequence of n strings, the task is to check if any two similar words come together and then destroy each other then print the number of words left in the sequence after this pairwise destruction.

**Examples:**

InputÂ : ab aa aa bcd ab

OutputÂ : 3

*As aa, aa destroys each other so,*

*ab bcd ab is the new sequence.*

Input : Â tom jerry jerry tom

Output :Â 0

*As first both jerry will destroy each other.*

*Then sequence will be tom, tom they will also destroy*

*each other. So, the final sequence doesnâ€™t contain any*

*word.*

</aside>

```python
def count_words(sequence):
    """
    Counts the number of unique words in a sequence of strings using a stack.
  
    Args:
      sequence (str): The sequence of strings.
  
    Returns:
      int: The number of unique words.
    """
  
    # Initialize an empty stack.
    stack = []
  
    # Split the sequence of strings by space and store it in a list.
    words = sequence.split()
  
    # Loop through each word in the list.
    for word in words:
  
      # Check if the stack is not empty and the word is equal to the top element of the stack.
      if stack and word == stack[-1]:
  
        # Pop the top element from the stack.
        stack.pop()
  
      # Otherwise, push the word into the stack.
      else:
        stack.append(word)
  
    # Return the size of the stack.
    return len(stack)

# test the code with an example input
sequence = "tom jerry jerry tom"
print(count_words(sequence))
```

ðŸ’¡ **Question 8**

Given an array of integers, the task is to find the maximum absolute difference between the nearest left and the right smaller element of every element in the array.

**Note:**Â If there is no smaller element on right side or left side of any element then we take zero as the smaller element. For example for the leftmost element, the nearest smaller element on the left side is considered as 0. Similarly, for rightmost elements, the smaller element on the right side is considered as 0.

**Examples:**
Input : arr[] = {2, 1, 8}
Output : 1
Left smaller  LS[] {0, 0, 1}
Right smaller RS[] {1, 0, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 1

Input  : arr[] = {2, 4, 8, 7, 7, 9, 3}
Output : 4
Left smaller   LS[] = {0, 2, 4, 4, 4, 7, 2}
Right smaller  RS[] = {0, 3, 7, 3, 3, 3, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 7 - 3 = 4

Input : arr[] = {5, 1, 9, 2, 5, 1, 7}
Output : 1
</aside>

```python
def max_abs_diff(arr):
    """
    Finds the maximum absolute difference between the nearest left and the right smaller element of every element in the array.
  
    Args:
      arr (list): The array.
  
    Returns:
      int: The maximum difference.
    """
  
    n = len(arr)
  
    # Initialize two empty lists, left_smaller and right_smaller.
    left_smaller = [0] * n
    right_smaller = [0] * n
  
    # Initialize the left_smaller list using a stack.
    stack = []
    for i in range(n):
      while stack and stack[-1] >= arr[i]:
        stack.pop()
  
      if stack:
        left_smaller[i] = stack[-1]
  
      stack.append(arr[i])
  
    # Initialize the right_smaller list using a stack.
    stack.clear()
    for i in range(n - 1, -1, -1):
      while stack and stack[-1] >= arr[i]:
        stack.pop()
  
      if stack:
        right_smaller[i] = stack[-1]
  
      stack.append(arr[i])
  
    # Compute maximum absolute difference.
    max_diff = max(abs(left_smaller[i] - right_smaller[i]) for i in range(n))
  
    return max_diff



# Test the code with example inputs
arr1 = [2, 1, 8]
print(max_abs_diff(arr1)) 

arr2 = [2, 4, 8, 7, 7, 9, 3]
print(max_abs_diff(arr2))

arr3 = [5, 1, 9, 2, 5, 1, 7]
print(max_abs_diff(arr3)) 
```
