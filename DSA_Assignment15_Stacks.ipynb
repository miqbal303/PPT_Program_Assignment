Given an arrayÂ **arr[ ]**Â of sizeÂ **N**Â havingÂ elements, theÂ task is to find the next greater element for each element of the array in order of their appearance in the array.Next greater element of an element in the array is the nearest element on the right which is greater than the current element.If there does not exist next greater of current element, then next greater element for current element is -1. For example, next greater of the last element is always -1.

**Example 1:**

```
Input:
N = 4, arr[] = [1 3 2 4]
Output:
3 4 4 -1
Explanation:
In the array, the next larger element
to 1 is 3 , 3 is 4 , 2 is 4 and for 4 ?
since it doesn't exist, it is -1.

```

**Example 2:**
Input:
N = 5, arr[] [6 8 0 1 3]
Output:
8 -1 1 3 -1
Explanation:
In the array, the next larger element to
6 is 8, for 8 there is no larger elements
hence it is -1, for 0 it is 1 , for 1 it
is 3 and then for 3 there is no larger
element on right and hence -1.
</aside>

````python
def nextGreater(arr, n):
    """
    Finds the next greater element for each element of the array.
  
    Args:
      arr (list): The array containing the elements.
      n (int): The number of elements in the array.
  
    Returns:
      list: The array containing the next greater elements.
    """
  
    # Initialize a stack to store the elements that have been seen so far.
    stack = []
  
    # Initialize an array to store the results.
    result = [-1] * n
  
    # Loop through the array from right to left.
    for i in range(n - 1, -1, -1):
      # Pop the elements from the stack that are smaller than the current element.
      while stack and stack[-1] <= arr[i]:
        stack.pop()
  
      # If the stack is not empty, the top element is the next greater element
      if stack:
        result[i] = stack[-1]
  
      # Push the current element to the stack
      stack.append(arr[i])
  
    # Return the result array
    return result

n = 5
arr = [6,8,0,1,3]
print(nextGreater(arr, n))
````

Given an arrayÂ **a**Â of integers of lengthÂ **n**, find the nearest smaller number for every element such that the smaller element is on left side.If no small element present on the left print -1.

**Example 1:**

```
Input: n = 3
a = {1, 6, 2}
Output: -1 1 1
Explaination: There is no number at the
left of 1. Smaller number than 6 and 2 is 1.
```

**Example 2:**
Input: n = 6
a = {1, 5, 0, 3, 4, 5}
Output: -1 1 -1 0 3 4
Explaination: Upto 3 it is easy to see
the smaller numbers. But for 4 the smaller
numbers are 1, 0 and 3. But among them 3
is closest. Similary for 5 it is 4.
</aside>

````python
def nearestSmaller(a, n):
    """
    Finds the nearest smaller number for every element in the array.
  
    Args:
      a (list): The array containing the elements.
      n (int): The number of elements in the array.
  
    Returns:
      list: The array containing the nearest smaller elements.
    """
  
    # Initialize a stack to store the elements that have been seen so far.
    stack = []
  
    # Initialize an array to store the results.
    result = []
  
    # Loop through the array from left to right.
    for i in range(n):
  
      # Pop the elements from the stack that are greater than or equal to the current element.
      while stack and stack[-1] >= a[i]:
        stack.pop()
  
      # If the stack is empty, there is no smaller element on the left.
      if not stack:
        result.append(-1)
  
      # If the stack is not empty, the top element is the nearest smaller element on the left.
      else:
        result.append(stack[-1])
  
      # Push the current element to the stack.
      stack.append(a[i])
  
    # Return the result array.
    return result


n = 6
a = [1, 5, 0, 3, 4, 5]

print(nearestSmaller(a, n))
````

Implement a StackÂ using two queuesÂ **q1**Â andÂ **q2**.

**Example 1:**

```

push(2)
push(3)
pop()
push(4)

Output:3 4
Explanation:
push(2)Â the stackÂ will be {2}
push(3)Â the stackÂ will be {2 3}
pop()   poped element will be 3Â the
Â        stackÂ will be {2}
push(4)Â the stackÂ will be {2 4}
pop() Â  poped element will be 4

```

**Example 2:**
Input:
push(2)
pop()
pop()
push(3)
Output:2 -1
</aside>

````python
class Stack:

    """
    A stack implemented using two queues.

    Attributes:
        q1 (list): The first queue.
        q2 (list): The second queue.
    """

    def __init__(self):
        """
        Initializes the stack.
        """
        self.q1 = []
        self.q2 = []

    def push(self, x):
        """
        Pushes an element onto the stack.

        Args:
            x (object): The element to push.
        """
        self.q1.append(x)

    def pop(self):
        """
        Pops an element from the stack.

        Returns:
            object: The popped element.
        """
        if not self.q1:
            return -1

        while len(self.q1) > 1:
            x = self.q1.pop(0)
            self.q2.append(x)

        x = self.q1.pop(0)
        self.q1, self.q2 = self.q2, self.q1
        return x

    
    
def main():
    stack = Stack()
    stack.push(2)
    stack.push(3)
    print(stack.pop())
    stack.push(4)
    print(stack.pop())


if __name__ == "__main__":
    main()
````

You are given a stackÂ **St**. You have to reverse the stack using recursion.

**Example 1:**

```
Input:St = {3,2,1,7,6}
Output:{6,7,1,2,3}
```

**Example 2:**
Input:St = {4,3,9,6}
Output:{6,9,3,4}
</aside>

````python
def reverse_stack(stack):
    """
    Reverses the stack using recursion.
  
    Args:
      stack (list): The stack to be reversed.
  
    Returns:
      list: The reversed stack.
    """
  
    if not stack:
      return stack
  
    top_element = stack.pop()
    reversed_stack = reverse_stack(stack)
    reversed_stack.append(top_element)
    return reversed_stack


def main():
    stack = [3, 2, 1, 7, 6]
    reversed_stack = reverse_stack(stack)
    print("Reversed stack:", reversed_stack)


if __name__ == "__main__":
  main()
````

ðŸ’¡ **Question 5**

You are given a stringÂ **S**, the task is to reverse the string using stack.

**Example 1:**
Input: S="GeeksforGeeks"
Output: skeeGrofskeeG
</aside>

````python
def reverse_string(s):
    """
    Reverses a string using a stack.
  
    Args:
      s (str): The string to be reversed.
  
    Returns:
      str: The reversed string.
    """
  
    stack = []
    for char in s:
      stack.append(char)
  
    reversed_string = ""
    while stack:
      reversed_string += stack.pop()
  
    return reversed_string


def main():
    s = "GeeksforGeeks"
    reversed_string = reverse_string(s)
    print("Reversed string:", reversed_string)


if __name__ == "__main__":
    main()
````

Given stringÂ **S**Â representing aÂ postfix expression, the task is to evaluate the expression and find the final value. Operators will only include the basic arithmetic operators likeÂ ***, /, + and -**.

**Example 1:**

```
Input: S = "231*+9-"
Output: -4
Explanation:
After solving the given expression,
we have -4 as result.

```

**Example 2:**
Input: S = "123+*8-"
Output: -3
Explanation:
After solving the given postfix
expression, we have -3 as result.
</aside>

````python
def evaluate_postfix(s):
    """
    Evaluates a postfix expression and returns the result.
  
    Args:
      s (str): The postfix expression.
  
    Returns:
      int: The result of the expression.
    """
  
    stack = []
    for ch in s:
      if ch in "+-*/":
        operand2 = stack.pop()
        operand1 = stack.pop()
        result = perform_operation(ch, operand1, operand2)
        stack.append(result)
      else:
        stack.append(int(ch))
  
    return stack.pop()


def perform_operation(ch, operand1, operand2):
    """
    Performs the operation specified by the character ch on operand1 and operand2.
  
    Args:
      ch (str): The operation character.
      operand1 (int): The first operand.
      operand2 (int): The second operand.
  
    Returns:
      int: The result of the operation.
    """
  
    if ch == "+":
      return operand1 + operand2
    elif ch == "-":
      return operand1 - operand2
    elif ch == "*":
      return operand1 * operand2
    else:
      return operand1 / operand2


def main():
    s = "231*+9-"
    result = evaluate_postfix(s)
    print("Result:", result)


if __name__ == "__main__":
  main()
````

ðŸ’¡ **Question 7**

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement theÂ `MinStack`Â class:

- `MinStack()`Â initializes the stack object.
- `void push(int val)`Â pushes the elementÂ `val`Â onto the stack.
- `void pop()`Â removes the element on the top of the stack.
- `int top()`Â gets the top element of the stack.
- `int getMin()`Â retrieves the minimum element in the stack.

You must implement a solution withÂ `O(1)`Â time complexity for each function.

**Example 1:**
Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
</aside>

````python
class MinStack:

    def __init__(self):
      """
      Initialize the stack object.
      """
      self.stack = []
      self.min_stack = []
  
    def push(self, val):
      """
      Pushes the element val onto the stack.
  
      Args:
        val (int): The element to push.
      """
      self.stack.append(val)
      if not self.min_stack or val <= self.min_stack[-1]:
        self.min_stack.append(val)
  
    def pop(self):
      """
      Removes the element on the top of the stack.
      """
      if self.stack:
        top_element = self.stack.pop()
        if top_element == self.min_stack[-1]:
          self.min_stack.pop()
  
    def top(self):
      """
      Gets the top element of the stack.
      """
      if self.stack:
        return self.stack[-1]
      else:
        return None
  
    def getMin(self):
      """
      Retrieve the minimum element in the stack.
      """
      if self.min_stack:
        return self.min_stack[-1]
      else:
        return None
  

def main():
    min_stack = MinStack()
    min_stack.push(-2)
    min_stack.push(0)
    min_stack.push(-3)
    print(min_stack.getMin())  # -3
    min_stack.pop()
    print(min_stack.top())  # 0
    print(min_stack.getMin())  # -2


if __name__ == "__main__":
    main()
````

GivenÂ `n`Â non-negative integers representing an elevation map where the width of each bar isÂ `1`, compute how much water it can trap after raining.

**Example 1**
<img src='https://pwskills.notion.site/image/https%3A%2F%2Fassets.leetcode.com%2Fuploads%2F2018%2F10%2F22%2Frainwatertrap.png?id=f0b192f5-9efa-4339-a63f-1a3a18daea19&table=block&spaceId=6fae2e0f-dedc-48e9-bc59-af2654c78209&width=820&userId=&cache=v2'>
Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.
    
    
    
Example 2:  
Input: height = [4,2,0,3,2,5]
Output: 9   
</aside>

````python
def trap(height):
    """
    Calculates the amount of water that can be trapped after raining.
  
    Args:
      height (list): The elevation map.
  
    Returns:
      int: The amount of water that can be trapped.
    """

    # initialize left and right pointers
    left = 0
    right = len(height) - 1
    # initialize left and right max heights
    left_max = 0
    right_max = 0
    # initialize the result
    result = 0
    # loop until left and right pointers meet
    while left < right:
        # update the left max height
        left_max = max(left_max, height[left])
        # update the right max height
        right_max = max(right_max, height[right])
        # if left max is smaller than right max
        if left_max < right_max:
            # add the difference between left max and current height to the result
            result += left_max - height[left]
            # move the left pointer to the right
            left += 1
        # otherwise
        else:
            # add the difference between right max and current height to the result
            result += right_max - height[right]
            # move the right pointer to the left
            right -= 1
    # return the result
    return result


def main():
    height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
    water = trap(height)
    print("Water trapped:", water)


if __name__ == "__main__":
    main()
````
